```python
# Linear and Logistic Regression from Scratch
## Introduction

In this assignment, linear regression and logistic regression are implemented from scratch using NumPy. The objective is to understand how these models learn from data using mathematical operations and optimization techniques.
```

```python
## Part 1.1: Linear Regression using Closed-Form Solution

#In this part, the weights are calculated directly using a mathematical formula without any iteration. This method finds the optimal solution in one step.

import numpy as np

# sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([50, 55, 65, 70, 80])

# add bias column
X = np.c_[np.ones((X.shape[0], 1)), X]

# closed-form solution
weights_cf = np.linalg.inv(X.T @ X) @ X.T @ y

print("Closed-form weights:", weights_cf)
```

```python
## Part 1.2: Linear Regression using Gradient Descent

#In this part, the model learns the weights step by step by reducing the error using gradient descent.

# initialize weights
weights_gd = np.zeros(X.shape[1])

learning_rate = 0.01
epochs = 1000
n = len(y)

for i in range(epochs):
    
    # predictions
    y_pred = X @ weights_gd
    
    # error
    error = y_pred - y
    
    # gradient
    gradient = (1/n) * (X.T @ error)
    
    # update weights
    weights_gd = weights_gd - learning_rate * gradient

print("Gradient descent weights:", weights_gd)
```

```python
## Part 2: Logistic Regression using Gradient Descent

In this part, logistic regression is implemented for binary classification. The sigmoid function is used to convert outputs into probabilities.
```

```python
# sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# sample classification data
X_clf = np.array([[1], [2], [3], [4], [5]])
y_clf = np.array([0, 0, 0, 1, 1])

# add bias
X_clf = np.c_[np.ones((X_clf.shape[0], 1)), X_clf]

# initialize weights
weights_log = np.zeros(X_clf.shape[1])

learning_rate = 0.01
epochs = 1000
n = len(y_clf)

for i in range(epochs):
    
    # linear output
    z = X_clf @ weights_log
    
    # probability
    preds = sigmoid(z)
    
    # error
    error = preds - y_clf
    
    # gradient
    gradient = (1/n) * (X_clf.T @ error)
    
    # update
    weights_log = weights_log - learning_rate * gradient

print("Logistic regression weights:", weights_log)
```

```python
## Part 3: Model Evaluation
##The models are evaluated using a train-test split. Mean squared error is used for linear regression, and accuracy is used for logistic regression.
# split data
split = int(0.8 * len(X))

X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# linear regression evaluation
weights_eval = np.linalg.inv(X_train.T @ X_train) @ X_train.T @ y_train
y_pred_test = X_test @ weights_eval

mse = np.mean((y_test - y_pred_test)**2)
print("Linear Regression MSE:",round(mse, 2))

# logistic regression evaluation
split = int(0.8 * len(X_clf))

X_train_clf, X_test_clf = X_clf[:split], X_clf[split:]
y_train_clf, y_test_clf = y_clf[:split], y_clf[split:]

weights_eval_log = np.zeros(X_train_clf.shape[1])

for i in range(epochs):
    preds = sigmoid(X_train_clf @ weights_eval_log)
    error = preds - y_train_clf
    gradient = (1/len(y_train_clf)) * (X_train_clf.T @ error)
    weights_eval_log = weights_eval_log - learning_rate * gradient

# predictions
probs = sigmoid(X_test_clf @ weights_eval_log)
pred_labels = (probs >= 0.5).astype(int)

accuracy = np.mean(pred_labels == y_test_clf)
print("Logistic Regression Accuracy:", accuracy)
```

```python
## Results and Analysis

The linear regression model produced predictions that were close to the actual values, resulting in a low mean squared error. This indicates that the model was able to capture the relationship between the input and output effectively.

The gradient descent approach showed steady improvement in the model parameters over iterations, leading to better predictions.

The logistic regression model was able to classify the data correctly with good accuracy. The use of the sigmoid function allowed the model to output probabilities and perform classification.

Overall, both models performed as expected for their respective tasks.
```
